```python
import csv
import pandas as pd
import numpy as np
df = pd.read_csv("./uiuc-students-by-curriculum.csv")
```

```python
def stringConvert(total):
    standin = total
    if (isinstance(standin, str)):
        standin = standin.replace(',', '')
        standin = (int)(standin)
    if (not np.isnan(standin)):
        standin = int(standin)
    if (isinstance(standin, str)):
        print(standin)
    return standin

def absoluteDifference(totals):
    print(totals)

def removeNan(specific):
    if (specific["MajorName"] != specific["MajorName"]):
        return specific["College"]
    return specific["MajorName"]

df2 = df.query("Fall == '1980' or Fall == '2018'")
df2 = df2[["Fall", "College", "Major Name", "Total", "Male", "Female"]]
df2["Total"] = df2["Total"].apply(stringConvert)
df2["Male"] = df2["Male"].apply(stringConvert)
df2["Female"] = df2["Female"].apply(stringConvert)
df2.rename(columns={'Major Name': 'MajorName'}, inplace=True)
df2["MajorName"] = df2.apply(removeNan, axis = 1)
```

```python
aggregatedMajor = df2.groupby(["Fall", "MajorName", "College"]).agg('sum')
aggregatedMajor = aggregatedMajor.sort_values(by=['College'])
aggregatedMajor.reset_index(inplace=True)  
```

```python
def absoluteDifference(totals, modifier):
    major = totals["MajorName"]
    if totals["Fall"] == 1980:
        two = aggregatedMajor[aggregatedMajor["Fall"] == 2018]
        if two[two["MajorName"] == major].empty:
            return -1*totals[modifier]
        else:
            return None
            
    nine = aggregatedMajor[aggregatedMajor["Fall"] == 1980]
    if nine[nine["MajorName"] == major].empty:
        return totals[modifier]
    else:
        rower = nine[nine["MajorName"] == major]
        return(totals[modifier] - rower[modifier].values[0])
    
two = aggregatedMajor[aggregatedMajor["Fall"] == 2018]
nine = aggregatedMajor[aggregatedMajor["Fall"] == 1980]
aggregatedMajor["AbsDif"] = aggregatedMajor.apply(lambda x: absoluteDifference(x, "Total"), axis = 1)
aggregatedMajor["AbsDifMale"] = aggregatedMajor.apply(lambda x: absoluteDifference(x, "Male"), axis = 1)
aggregatedMajor["AbsDifFemale"] = aggregatedMajor.apply(lambda x: absoluteDifference(x, "Female"), axis = 1)
aggregatedMajor
```

```python
def relativeDifference(totals, modifier):
    major = totals["MajorName"]
    if totals["Fall"] == 1980:
        two = aggregatedMajor[aggregatedMajor["Fall"] == 2018]
        if two[two["MajorName"] == major].empty:
            return -100.0
        else:
            return None
            
    nine = aggregatedMajor[aggregatedMajor["Fall"] == 1980]
    rower = nine[nine["MajorName"] == major]
    if nine[nine["MajorName"] == major].empty or rower[modifier].values[0] == 0:
        return 10000.0
    else:
        rower = nine[nine["MajorName"] == major]
        return((totals[modifier] - rower[modifier].values[0]) *1.0 / rower[modifier].values[0]*100)
    
aggregatedMajor["RelDif"] = aggregatedMajor.apply(lambda x: relativeDifference(x, "Total"), axis = 1)
aggregatedMajor["RelDifMale"] = aggregatedMajor.apply(lambda x: relativeDifference(x, "Male"), axis = 1)
aggregatedMajor["RelDifFemale"] = aggregatedMajor.apply(lambda x: relativeDifference(x, "Female"), axis = 1)
aggregatedMajor = aggregatedMajor[aggregatedMajor["AbsDif"] == aggregatedMajor["AbsDif"]]
aggregatedMajor
```
